```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Inline indicates graphs should be presented as a cell output
%matplotlib inline
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile
selected_manufacturers = ['Audi','Hyundai','Suzuki','Toyota']
filtered_data = car_data[car_data['Manufacturer'].isin(selected_manufacturers)]
plt.figure(figsize=(10,6))
ax = sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=filtered_data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, horizontalalignment='right')

plt.xlabel('Manufacturer')
plt.ylabel('Revs per mile')
plt.title('Boxplot for Revs per mile by Manufacturer')
plt.tight_layout()
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?



**Boxplot answer:**
Suzuki has the highest revs per mile which is a above 3300 and under 3400. The bar plot shows it as a horizontal line which indicates most likely the data in dataset for 'Suzuki' is on lesser side.

### 2. Histogram

```python
# Histogram of MPG
plt.subplot(1, 2, 1)
plt.hist(car_data['MPG.city'],bins=20,color='blue',alpha=0.7)

plt.subplot(1,2,1)
plt.hist(car_data['MPG.highway'],bins=20,color='green',alpha=0.7)
plt.xlabel('MPG in city(blue)/highway(green)')
plt.ylabel('Frequency')
plt.title('Histogram of MPG in city/highway')

plt.tight_layout()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**

Driving in highway looks more efficient in terms of fuel efficiency than city for this dataset. The MPG City Histogram is more skewed to the right whereas the MPG Highway Histogram is more normal.This indicates that the city histogram has more values near the lower end of the range and higher values are infrequent. Hence driving in highway is more efficient in this data set.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.plot(car_data['Wheelbase'], car_data['Turn.circle'],color='blue')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.title('Line plot of Wheelbase vs Turning Circle')
plt.grid(True)
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  


In general, there is a positive correlation between Wheelbase and Turning circle. As Wheelbase increases Turning circle also increases.

### 4. Barplot

```python
# Barplot for mean horsepower
mean_horsepower_by_type = car_data.groupby('Type')['Horsepower'].mean()
mean_horsepower_by_type.plot(kind='bar',color='blue')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.title('Mean Horsepower by Car Type')
plt.tight_layout()
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**

Yes the bar plot does indicate that a larger car has a larger mean horsepower.So in this dataset a larger car means larger horsepower. It is around 175. This is followed by mid size cars which is above 150 and little below 175.
