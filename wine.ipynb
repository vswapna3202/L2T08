## **Compulsary Task 2** 
### **1. Multi-grid plot**

1. Read the wine.csv file

```python
# Load data
wine_data = pd.read_csv('wine.csv') 
```

2 .Filter the dataset to only contain “Cabernet Sauvignon”, “Pinot
Noir”, and “Chardonnay” wines. Use the variety column.

```python
selected_varieties = ['Cabernet Sauvignon', 'Pinot Noir', 'Chardonnay']
filtered_data = wine_data[wine_data['variety'].isin(selected_varieties)]
points_column = filtered_data[['points','variety']]
```

3. For each of these three wine varieties, show a histogram plot of the
“points” column.


```python
grid = sns.FacetGrid(points_column, col="variety",col_wrap=3, height=5)
grid.map(plt.hist, "points", bins=20, color="blue",alpha=0.7, edgecolor="gold");
grid.set_titles("{col_name}",fontsize=14)
grid.set_axis_labels('Points', 'Variety')
plt.subplots_adjust(
    top=0.9,  
    bottom=0.1,  
    left=0.1,  
    right=0.9,  
    wspace=0.4,  
    hspace=0.4  
)
grid.fig.suptitle('Points vs Variety', fontsize=14)

# Show the plot
plt.show()
```
